```python
import pandas as pd

df = pd.read_csv("../postprocessed/mediterranean lightnings 2009-2021.csv")

df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')

df['Month'] = df['Date'].apply(lambda x: x.strftime('%Y-%m'))

df = df.drop(df.columns[[0,1]], axis=1)

new_order = ['Month', 'Latitude', 'Longitude', 'Power (J)']
df = df.reindex(columns=new_order)

df['Latitude'] = df['Latitude'].round(1)
df['Longitude'] = df['Longitude'].round(1)

df = df.sort_values(['Month', 'Latitude', 'Longitude'])
df = df.reset_index()
merged_df = df.drop(df.columns[0], axis=1)
merged_df = merged_df.drop(merged_df[(merged_df.Month == '2021-12') | (merged_df.Month == '2022-01')].index)
```

```python
import pandas as pd 

salinity_df = pd.read_csv("../postprocessed/Salinity 2009-2021.csv")

salinity_df = salinity_df[salinity_df['Longitude'] >= -5]

salinity_df['Latitude'] = salinity_df['Latitude'].round(1)
salinity_df['Longitude'] = salinity_df['Longitude'].round(1)


salinity_df = salinity_df.groupby(['Month','Latitude', 'Longitude'], as_index=False).agg({'Sodium': 'mean'})

merged_df = merged_df.merge(salinity_df, on=["Month", "Latitude", "Longitude"], how='left')
```

```python
import pandas as pd 

temperature_df = pd.read_csv("../postprocessed/Temperature 2009-2021.csv")

temperature_df = temperature_df[temperature_df['Longitude'] >= -5]

temperature_df['Latitude'] = temperature_df['Latitude'].round(1)
temperature_df['Longitude'] = temperature_df['Longitude'].round(1)


temperature_df = temperature_df.groupby(['Month','Latitude', 'Longitude'], as_index=False).agg({'Temperature': 'mean'})

merged_df = merged_df.merge(temperature_df, on=["Month", "Latitude", "Longitude"], how='left')
```
