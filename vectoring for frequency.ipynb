```python
import pandas as pd
import numpy as np

df = pd.read_csv("../postprocessed/mediterranean lightnings 2009-2021.csv")

df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')

df['Month'] = df['Date'].apply(lambda x: x.strftime('%Y-%m'))

df['Latitude'] = np.floor(df['Latitude'])
df['Longitude'] = np.floor(df['Longitude'])

df = df.groupby(['Month','Latitude', 'Longitude'], as_index=False).agg({'Power (J)': ['count', "mean"]})

df.columns = df.columns.map(''.join) # Flatten column names
#df = df.reset_index(col_level=1)

df
```

```python
merged_df=df.copy()
merged_df
```

```python
import pandas as pd

salinity_df = pd.read_csv("../postprocessed/Salinity 2009-2021.csv")

salinity_df = salinity_df[salinity_df['Longitude'] >= -5]

salinity_df['Latitude'] = np.floor(salinity_df['Latitude'])
salinity_df['Longitude'] = np.floor(salinity_df['Longitude'])

salinity_df = salinity_df.groupby(['Month','Latitude', 'Longitude'], as_index=False).agg({'Sodium': 'mean'})

merged_df = merged_df.merge(salinity_df, on=["Month", "Latitude", "Longitude"], how='right')
```
