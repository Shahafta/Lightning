```python
import os
import pandas as pd

folder_path = "../old uncertain cs"

total_rows = 0

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"): # assuming all files are CSVs
        df = pd.read_csv(os.path.join(folder_path, filename))
        total_rows += len(df)

print("Total rows:", total_rows)
```
