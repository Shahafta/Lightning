```python
import pandas as pd
import numpy as np

df = pd.read_csv("../postprocessed/mediterranean lightnings 2009-2021.csv")

df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')

df['Month'] = df['Date'].apply(lambda x: x.strftime('%Y-%m'))

df['Latitude'] = np.floor(df['Latitude'])
df['Longitude'] = np.floor(df['Longitude'])
```

```python
df2 = df.groupby(['Month'], as_index=False).agg({'Power (J)': 'count'})

df2 = df2.rename(columns={"Power (J)":"Count"})

df2
```

```python
df3 = df.groupby(['Month'], as_index=False).agg({'Power (J)': 'mean'})

df3 = df3.rename(columns={"Power (J)":"Mean"})

df3
```

```python
merged_df = df2.merge(df3, on=["Month"], how='left')
merged_df
```
