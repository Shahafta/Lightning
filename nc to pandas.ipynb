```python
import pandas as pd
import xarray as xr
import os

folder_path = '../DAILY REST/'

# iterate over all files in the folder
for file_name in os.listdir(folder_path):
    # check if the current item is a file
    if os.path.isfile(os.path.join(folder_path, file_name)):
        # do something with the file
        ds = xr.open_dataset(os.path.join(folder_path, file_name))
        df = ds.to_dataframe()
        ds.close()
        df = df.reset_index()
        df.drop(columns=['depth'], inplace=True)
        df = df.dropna()
        df['time'] = df['time'].astype(str)
        #df = df[(df['time'].str[5:7]=="12")|(df['time'].str[5:7]=="01")|(df['time'].str[5:7]=="02")]
        concat = pd.concat([concat, df], ignore_index=True)
        df = None
```

```python
concat = concat.sort_values(['time', 'latitude','longitude'])
concat
```

```python
concat.to_csv("../daily rest.csv", index=False)
```

```python
import xarray as xr
import pandas as pd

# Open the .nc file
ds = xr.open_dataset('2009-01.nc')

# Convert the xarray dataset to a Pandas dataframe
df = ds.to_dataframe()

# Close the file
ds.close()

df = df.reset_index()

df.drop(columns=['depth'], inplace=True)

df = df.dropna(subset=['so'])

df['Month'] = df['time'].astype(str)
df['Month'] = df['Month'].str[5:7]
df = df[(df['Month']=='12')|(df['Month']=='01')|(df['Month']=='02')]
df['Month'] = df['time'].astype(str)
df['Month'] = df['Month'].str[:7]
df.drop(columns=['time'], inplace=True)
df = df.rename(columns={'lon': 'Longitude', 'lat': 'Latitude','so':"Sodium"})
df = df.reindex(columns=['Latitude', 'Longitude',"Month","Sodium"])
df['Latitude'] = df['Latitude'].round(3)
df['Longitude'] = df['Longitude'].round(3)

df
```

```python
import xarray as xr
import pandas as pd
from global_land_mask import globe
import pandas as pd
import math
from os import listdir
from os.path import isfile, join

final = None
# Open the .nc file
files = [f for f in listdir("F:/Datasets/percipitation/") if isfile(join("F:/Datasets/percipitation/", f))]

for x in range(len(files)):
    ds = xr.open_dataset('../percipitation/'+str(files[x]))

    # Convert the xarray dataset to a Pandas dataframe
    df = ds.to_dataframe()

    # Close the file
    ds.close()

    df = df.reset_index()

    df.drop(columns=['nv','lat_bounds','lon_bounds','time_bounds'], inplace=True)

    df['Month'] = df['time'].astype(str)
    df['Month'] = df['Month'].str[5:7]
    df = df[(df['Month']=='12')|(df['Month']=='01')|(df['Month']=='02')]
    df['Month'] = df['time'].astype(str)
    df['Month'] = df['Month'].str[:7]
    df.drop(columns=['time'], inplace=True)
    df['precip']=df['precip']-df['precip_error']
    df = df.rename(columns={'longitude': 'Longitude', 'latitude': 'Latitude','precip':"Precipitation"})
    df = df.reindex(columns=['Latitude', 'Longitude',"Month","Precipitation"])
    df['Latitude'] = df['Latitude'].round(3)
    df['Longitude'] = df['Longitude'].round(3)

    final = pd.concat([final,df],ignore_index=True)

final
```

```python
final.loc[:, "Longitude"] = final["Longitude"].apply(lambda x: x - 180)
final
```

```python
final.to_csv("../Precipitation 2009-2020.csv",index=False)
```

```python
import xarray as xr
import pandas as pd
from global_land_mask import globe
import pandas as pd
import math
from os import listdir
from os.path import isfile, join

final = None
# Open the .nc file
files = [f for f in listdir("F:/Datasets/PRECIPITATION/") if isfile(join("F:/Datasets/PRECIPITATION/", f))]

for x in range(len(files)):
    ds = xr.open_dataset('../PRECIPITATION/'+str(files[x]))

    # Convert the xarray dataset to a Pandas dataframe
    df = ds.to_dataframe()

    # Close the file
    ds.close()

    df = df.reset_index()

    df=df[df["nv"]==0]
    df.drop(columns=['nv','lat_bounds','lon_bounds','time_bounds'], inplace=True)

    final = pd.concat([final,df],ignore_index=True)

final
```
